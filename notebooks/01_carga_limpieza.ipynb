```python
# PARA CARGAR CSV O EXCEL

def cargar_archivo(ruta):
    ruta = Path(ruta)

    if ruta.suffix.lower() == ".csv":
        try:
            return pd.read_csv(ruta, encoding="utf-8", sep=None, engine="python")
        except UnicodeDecodeError:
            return pd.read_csv(ruta, encoding="latin1", sep=None, engine="python")
        

    elif ruta.suffix.lower() in [".xlsx", ".xls"]:
        return pd.read_excel(ruta)
    
    else:
        raise ValueError(f"Formato no soportado: {ruta.suffix}")
```

```python
import re
import unicodedata
```

```python
# LIMPIEZA DE NOMBRES DE COLUMNAS

def limpiar_texto_columna(texto):
    texto = str(texto)
    texto = texto.replace("\ufeff", "")
    texto = texto.strip().lower()
    texto = unicodedata.normalize("NFKD", texto).encode("ascii", "ignore").decode("utf-8")
    texto = re.sub(r"[^a-z0-9]+", "_", texto)
    texto = texto.strip("-")
    return texto

def limpiar_columnas(df):
    df = df.copy()
    df.columns = [limpiar_texto_columna(col) for col in df.columns]
    return df
```

```python
siniestros = limpiar_columnas(siniestros)
vehiculos = limpiar_columnas(vehiculos)
precipitacion = limpiar_columnas(precipitacion)
temperatura = limpiar_columnas(temperatura)
fallecidos = limpiar_columnas(fallecidos)
poblacion = limpiar_columnas(poblacion)

datasets = {
    "siniestros": siniestros,
    "vehiculos": vehiculos,
    "precipitacion": precipitacion,
    "temperatura": temperatura,
    "fallecidos": fallecidos,
    "poblacion": poblacion
}

for nombre, df in datasets.items():
    print("=" * 80)
    print(nombre.upper())
    print(df.columns.tolist())
```

```python
# LIMPIAR TEXTOS Y NUMEROS
def limpiar_texto_valor(valor):
    if pd.isna(valor):
        return np.nan
    
    valor = str(valor)
    valor = valor.replace("\ufeff", "")
    valor = valor.strip().upper()
    valor = unicodedata.normalize("NFKD", valor).encode("ascii", "ignore").decode("utf-8")
    valor = re.sub(r"\s+", " ", valor)
    
    return valor


def convertir_numero_columna(serie):
    def convertir(valor):
        if pd.isna(valor):
            return np.nan
        
        if isinstance(valor, (int, float, np.integer, np.floating)):
            return valor
        
        valor = str(valor).strip()
        valor = valor.replace("\ufeff", "")
        valor = re.sub(r"[^0-9,.\-]", "", valor)
        
        if valor == "":
            return np.nan
        
        if "," in valor and "." in valor:
            if valor.rfind(",") > valor.rfind("."):
                valor = valor.replace(".", "").replace(",", ".")
            else:
                valor = valor.replace(",", "")
        
        elif "," in valor:
            partes = valor.split(",")
            if len(partes) == 2 and len(partes[1]) == 3:
                valor = valor.replace(",", "")
            else:
                valor = valor.replace(",", ".")
        
        elif valor.count(".") > 1:
            valor = valor.replace(".", "")
        
        return pd.to_numeric(valor, errors="coerce")
    
    return serie.apply(convertir)
```

```python
# LIMPIEZA DE SINIESTROS
# ==============================

# Limpiar columnas de texto
columnas_texto_siniestros = [
    "mes", "dia", "hora", "provincia", "canton", "zona", "clase", "causa"
]

for col in columnas_texto_siniestros:
    siniestros[col] = siniestros[col].apply(limpiar_texto_valor)


# Convertir columnas numéricas
columnas_numericas_siniestros = [
    "num_fallecido", "num_lesionado", "total_victimas"
]

for col in columnas_numericas_siniestros:
    siniestros[col] = convertir_numero_columna(siniestros[col])


# Convertir columnas numéricas a enteros permitiendo nulos
for col in columnas_numericas_siniestros:
    siniestros[col] = siniestros[col].round().astype("Int64")


# Convertir valores negativos en nulos
for col in columnas_numericas_siniestros:
    siniestros.loc[siniestros[col] < 0, col] = pd.NA


# Convertir mes a número
siniestros["mes_num"] = (
    siniestros["mes"]
    .apply(convertir_mes)
    .astype("Int64")
)


# Convertir día de semana a número
mapa_dias_semana = {
    "LUNES": 1,
    "MARTES": 2,
    "MIERCOLES": 3,
    "JUEVES": 4,
    "VIERNES": 5,
    "SABADO": 6,
    "DOMINGO": 7
}

siniestros["dia_semana_num"] = (
    siniestros["dia"]
    .map(mapa_dias_semana)
    .astype("Int64")
)


# Extraer hora inicial
def extraer_hora(valor):
    if pd.isna(valor):
        return np.nan

    texto = str(valor).strip()
    coincidencia = re.search(r"\d{1,2}", texto)

    if coincidencia:
        hora = int(coincidencia.group())

        if 0 <= hora <= 23:
            return hora

    return np.nan


siniestros["hora_num"] = (
    siniestros["hora"]
    .apply(extraer_hora)
    .astype("Int64")
)


# Clasificar franja horaria
def clasificar_franja_horaria(hora):
    if pd.isna(hora):
        return np.nan

    hora = int(hora)

    if 0 <= hora <= 5:
        return "MADRUGADA"
    elif 6 <= hora <= 11:
        return "MANANA"
    elif 12 <= hora <= 17:
        return "TARDE"
    elif 18 <= hora <= 23:
        return "NOCHE"

    return np.nan


siniestros["franja_horaria"] = (
    siniestros["hora_num"]
    .apply(clasificar_franja_horaria)
)


# Crear columna año porque el archivo corresponde a 2019
siniestros["anio"] = 2019


# Crear fecha mensual de referencia
# IMPORTANTE: dia es día de semana, no día del mes.
siniestros["fecha_mes"] = pd.to_datetime(
    {
        "year": siniestros["anio"],
        "month": siniestros["mes_num"],
        "day": 1
    },
    errors="coerce"
)


# Validar total de víctimas
siniestros["total_victimas_calculado"] = (
    siniestros["num_fallecido"].fillna(0)
    + siniestros["num_lesionado"].fillna(0)
).astype("Int64")


siniestros["total_victimas_inconsistente"] = (
    siniestros["total_victimas"].notna()
    & (
        siniestros["total_victimas"]
        != siniestros["total_victimas_calculado"]
    )
)


# Eliminar duplicados
duplicados_antes = siniestros.duplicated().sum()

siniestros = (
    siniestros
    .drop_duplicates()
    .reset_index(drop=True)
)

print("Duplicados eliminados:", duplicados_antes)


# Crear ID único
siniestros["id_siniestro"] = siniestros.index + 1


# Revisar resultado
siniestros[
    [
        "id_siniestro",
        "anio",
        "mes",
        "mes_num",
        "dia",
        "dia_semana_num",
        "hora",
        "hora_num",
        "franja_horaria",
        "fecha_mes",
        "provincia",
        "canton",
        "zona",
        "clase",
        "causa",
        "num_fallecido",
        "num_lesionado",
        "total_victimas",
        "total_victimas_calculado",
        "total_victimas_inconsistente"
    ]
].head()
```

```python
# LIMPIEZA DE VEHICULOS
# Limpieza de textos
for col in ["provincia", "marca", "clase", "combustible", "modelo", "servicio", "estratone", "estrapasajero"]:
    if col in vehiculos.columns:
        vehiculos[col] = vehiculos[col].apply(limpiar_texto_valor)

# Limpieza de números
for col in ["pasajeros", "tonelaje"]:
    if col in vehiculos.columns:
        vehiculos[col] = convertir_numero_columna(vehiculos[col])

vehiculos["anio"] = 2019

vehiculos.head()
```

```python
# LIMPIEZA DE FALLECIDOS

# Limpieza de textos
for col in ["genero_de_la_victima", "estado_civil_de_la_victima", "condicion", "tipo_de_accidente", "canton", "provincia"]:
    if col in fallecidos.columns:
        fallecidos[col] = fallecidos[col].apply(limpiar_texto_valor)

# Fechas
for col in ["fecha_de_muerte_victima", "fecha_siniestro", "fecha_de_creacion_de_proteccion_de_fallecidos"]:
    if col in fallecidos.columns:
        fallecidos[col] = pd.to_datetime(fallecidos[col], errors="coerce", dayfirst=True)

# Filtrar solo fallecidos con siniestro en 2019
fallecidos_2019 = fallecidos[fallecidos["fecha_siniestro"].dt.year == 2019].copy()

fallecidos_2019["anio"] = 2019
fallecidos_2019["mes_num"] = fallecidos_2019["fecha_siniestro"].dt.month

print("Filas originales SPPAT:", fallecidos.shape[0])
print("Filas SPPAT 2019:", fallecidos_2019.shape[0])

fallecidos_2019.head()
```

```python
# LIMPIAR CENSO
for col in ["provincia", "canton", "parroquia"]:
    if col in poblacion.columns:
        poblacion[col] = poblacion[col].apply(limpiar_texto_valor)

for col in ["poblacion", "superficie_de_la_parroquia_km2", "densidad_poblacional"]:
    if col in poblacion.columns:
        poblacion[col] = convertir_numero_columna(poblacion[col])

# Evitar posibles filas de totales si existieran
if "parroquia" in poblacion.columns:
    poblacion = poblacion[~poblacion["parroquia"].astype(str).str.contains("TOTAL", na=False)].copy()

poblacion.head()
```

```python
# TRANSFORMAR PRECIPITACION Y TEMPERATURA
meses_clima = ["ene", "feb", "mar", "abr", "may", "jun", "jul", "ago", "sep", "oct", "nov", "dic"]

def transformar_clima(df, nombre_variable):
    df = df.copy()
    
    # Limpiar textos
    for col in ["estacion", "nombreestacion"]:
        if col in df.columns:
            df[col] = df[col].apply(limpiar_texto_valor)
    
    # Convertir numéricos
    for col in ["longitud2", "latitud2", "altitud", "anio"]:
        if col in df.columns:
            df[col] = convertir_numero_columna(df[col])
    
    id_cols = [col for col in ["estacion", "nombreestacion", "longitud2", "latitud2", "altitud", "anio"] if col in df.columns]
    value_cols = [col for col in meses_clima if col in df.columns]
    
    df_largo = df.melt(
        id_vars=id_cols,
        value_vars=value_cols,
        var_name="mes",
        value_name=nombre_variable
    )
    
    df_largo["mes_num"] = df_largo["mes"].apply(convertir_mes)
    df_largo[nombre_variable] = convertir_numero_columna(df_largo[nombre_variable])
    
    return df_largo

precipitacion_larga = transformar_clima(precipitacion, "precipitacion")
temperatura_larga = transformar_clima(temperatura, "temperatura")

# Filtrar 2019 por seguridad
precipitacion_larga = precipitacion_larga[precipitacion_larga["anio"] == 2019].copy()
temperatura_larga = temperatura_larga[temperatura_larga["anio"] == 2019].copy()

display(precipitacion_larga.head())
display(temperatura_larga.head())
```
