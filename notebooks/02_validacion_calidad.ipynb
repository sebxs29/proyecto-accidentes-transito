```python
import pandas as pd
import numpy as np
from pathlib import Path
PROCESSED = Path("../data/processed")
DOCS = Path("../docs")

DOCS.mkdir(parents=True, exist_ok=True)

print("Carpeta processed:", PROCESSED.resolve())
print("Carpeta docs:", DOCS.resolve())
```

```python
# CARGAR LOS DATOS LIMPIOS
siniestros = pd.read_csv(PROCESSED / "siniestros_2019_limpio.csv")
vehiculos = pd.read_csv(PROCESSED / "vehiculos_matriculados_2019_limpio.csv")
precipitacion = pd.read_csv(PROCESSED / "precipitacion_2019_limpia.csv")
temperatura = pd.read_csv(PROCESSED / "temperatura_2019_limpia.csv")
fallecidos = pd.read_csv(PROCESSED / "fallecidos_sppat_2019_limpio.csv")
poblacion = pd.read_csv(PROCESSED / "poblacion_censo_2022_limpia.csv")
```

```python
datasets_limpios = {
    "siniestros_limpio": siniestros,
    "vehiculos_limpio": vehiculos,
    "precipitacion_limpia": precipitacion,
    "temperatura_limpia": temperatura,
    "fallecidos_sppat_2019_limpio": fallecidos,
    "poblacion_limpia": poblacion
}

for nombre, df in datasets_limpios.items():
    print("=" * 60)
    print(nombre)
    print(df.shape)
    display(df.head())
```

```python
# REPORTe DE CALIDAD DESPUES
def reporte_calidad(nombre, df):
    return {
        "dataset": nombre,
        "filas": df.shape[0],
        "columnas": df.shape[1],
        "duplicados": int(df.duplicated().sum()),
        "nulos_totales": int(df.isnull().sum().sum()),
        "columnas_con_nulos": int((df.isnull().sum() > 0).sum())
    }

calidad_despues = pd.DataFrame([
    reporte_calidad(nombre, df)
    for nombre, df in datasets_limpios.items()
])

calidad_despues
```

```python
calidad_despues.to_csv(
    DOCS / "reporte_calidad_despues.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Reporte de calidad después exportado.")
```

```python
#VALIDACIONES
validaciones = []

def agregar_validacion(dataset, validacion, condicion, detalle):
    validaciones.append({
        "dataset": dataset,
        "validacion": validacion,
        "resultado": "OK" if condicion else "REVISAR",
        "detalle": detalle
    })
```

```python
# SINIESTROS
agregar_validacion(
    "siniestros",
    "Tiene registros",
    siniestros.shape[0] > 0,
    f"Total registros: {siniestros.shape[0]}"
)

agregar_validacion(
    "siniestros",
    "Meses válidos",
    siniestros["mes_num"].between(1, 12).all(),
    f"Meses inválidos: {(~siniestros['mes_num'].between(1, 12)).sum()}"
)

agregar_validacion(
    "siniestros",
    "Días de semana válidos",
    siniestros["dia_semana_num"].between(1, 7).all(),
    f"Días inválidos: {(~siniestros['dia_semana_num'].between(1, 7)).sum()}"
)

agregar_validacion(
    "siniestros",
    "Horas válidas",
    siniestros["hora_num"].between(0, 23).all(),
    f"Horas inválidas: {(~siniestros['hora_num'].between(0, 23)).sum()}"
)

victimas_negativas = (
    (siniestros["num_fallecido"] < 0) |
    (siniestros["num_lesionado"] < 0) |
    (siniestros["total_victimas"] < 0)
).sum()

agregar_validacion(
    "siniestros",
    "Víctimas no negativas",
    victimas_negativas == 0,
    f"Filas con víctimas negativas: {victimas_negativas}"
)

inconsistencia_victimas = (
    siniestros["total_victimas"].fillna(0) !=
    siniestros["num_fallecido"].fillna(0) + siniestros["num_lesionado"].fillna(0)
).sum()

agregar_validacion(
    "siniestros",
    "Total víctimas consistente",
    inconsistencia_victimas == 0,
    f"Inconsistencias: {inconsistencia_victimas}"
)
```

```python
# VEHICULOS
agregar_validacion(
    "vehiculos",
    "Tiene registros",
    vehiculos.shape[0] > 0,
    f"Total registros: {vehiculos.shape[0]}"
)

agregar_validacion(
    "vehiculos",
    "Provincia no nula",
    vehiculos["provincia"].notna().all(),
    f"Provincias nulas: {vehiculos['provincia'].isna().sum()}"
)

agregar_validacion(
    "vehiculos",
    "Pasajeros no negativos",
    (vehiculos["pasajeros"].dropna() >= 0).all(),
    f"Pasajeros negativos: {(vehiculos['pasajeros'] < 0).sum()}"
)

agregar_validacion(
    "vehiculos",
    "Tonelaje no negativo",
    (vehiculos["tonelaje"].dropna() >= 0).all(),
    f"Tonelaje negativo: {(vehiculos['tonelaje'] < 0).sum()}"
)
```

```python
# CLIMA
agregar_validacion(
    "precipitacion",
    "Meses válidos",
    precipitacion["mes_num"].between(1, 12).all(),
    f"Meses inválidos: {(~precipitacion['mes_num'].between(1, 12)).sum()}"
)

agregar_validacion(
    "precipitacion",
    "Precipitación no negativa",
    (precipitacion["precipitacion"].dropna() >= 0).all(),
    f"Valores negativos: {(precipitacion['precipitacion'] < 0).sum()}"
)

agregar_validacion(
    "temperatura",
    "Meses válidos",
    temperatura["mes_num"].between(1, 12).all(),
    f"Meses inválidos: {(~temperatura['mes_num'].between(1, 12)).sum()}"
)

agregar_validacion(
    "temperatura",
    "Temperatura en rango razonable",
    temperatura["temperatura"].dropna().between(-20, 50).all(),
    f"Valores fuera de rango: {(~temperatura['temperatura'].dropna().between(-20, 50)).sum()}"
)
```

```python
# FALLECIDOS
fallecidos["fecha_siniestro"] = pd.to_datetime(
    fallecidos["fecha_siniestro"],
    errors="coerce"
)

agregar_validacion(
    "fallecidos_sppat",
    "Tiene registros",
    fallecidos.shape[0] > 0,
    f"Total registros: {fallecidos.shape[0]}"
)

anios_fallecidos = fallecidos["fecha_siniestro"].dt.year.dropna().unique()

agregar_validacion(
    "fallecidos_sppat",
    "Solo contiene año 2019",
    set(anios_fallecidos) == {2019},
    f"Años encontrados: {anios_fallecidos}"
)

agregar_validacion(
    "fallecidos_sppat",
    "Provincia no nula",
    fallecidos["provincia"].notna().all(),
    f"Provincias nulas: {fallecidos['provincia'].isna().sum()}"
)
```

```python
# POBLACION
agregar_validacion(
    "poblacion",
    "Tiene registros",
    poblacion.shape[0] > 0,
    f"Total registros: {poblacion.shape[0]}"
)

agregar_validacion(
    "poblacion",
    "Población positiva",
    (poblacion["poblacion"].dropna() > 0).all(),
    f"Población <= 0: {(poblacion['poblacion'] <= 0).sum()}"
)

agregar_validacion(
    "poblacion",
    "Provincia no nula",
    poblacion["provincia"].notna().all(),
    f"Provincias nulas: {poblacion['provincia'].isna().sum()}"
)
```

```python
# VALIDACION SI LAS PROVINCIAS DE SINIESTROS EXISTEN EN POBLACION
provincias_siniestros = set(siniestros["provincia"].dropna().unique())
provincias_poblacion = set(poblacion["provincia"].dropna().unique())

provincias_sin_match = (
    provincias_siniestros - provincias_poblacion
)

detalle_siniestros = (
    "Ninguna"
    if len(provincias_sin_match) == 0
    else ", ".join(sorted(provincias_sin_match))
)

agregar_validacion(
    "integridad_referencial",
    "Provincias de siniestros existen en población",
    len(provincias_sin_match) == 0,
    f"Provincias sin coincidencia: {detalle_siniestros}"
)
```

```python
# VALIDACIÓN: PROVINCIAS DE VEHÍCULOS EXISTEN EN POBLACIÓN
provincias_vehiculos = set(vehiculos["provincia"].dropna().unique())
provincias_vehiculos_sin_match = provincias_vehiculos - provincias_poblacion

detalle_vehiculos = (
    "Ninguna"
    if len(provincias_vehiculos_sin_match) == 0
    else ", ".join(sorted(provincias_vehiculos_sin_match))
)

agregar_validacion(
    "integridad_referencial",
    "Provincias de vehículos existen en población",
    len(provincias_vehiculos_sin_match) == 0,
    f"Provincias sin coincidencia: {detalle_vehiculos}"
)
```

```python
reporte_validaciones = pd.DataFrame(validaciones)

reporte_validaciones
```
